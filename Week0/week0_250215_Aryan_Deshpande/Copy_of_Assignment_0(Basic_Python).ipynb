**Q.4**
**Linear Regression:-** https://tahera-firdose.medium.com/linear-regression-derivation-of-slope-and-intercept-using-ordinary-least-square-971534ec6b77

(i) Given:

```python
X = [1,2,3,4,5]
y = [2,4,5,4,5]
```

Compute:

mean of X

mean of y

slope (m)

intercept (c)

predicted values

No libraries allowed.(sklearn or numpy)


```python
#mean of X
X = [1, 2, 3, 4, 5]
mean_value_x = sum(X) / len(X)
print(f"Mean = {mean_value_x}")
```

```python
#mean of Y
Y = [2,4,5,4,5]
mean_value_y = sum(Y) / len(Y)
print(f"Mean = {mean_value_y}")
```

```python
#slope m
def calculate_slope_ols(x_values, y_values):
    numerator = 0
    denominator = 0
    for x, y in zip(x_values, y_values):
        numerator += (x - mean_value_x) * (y - mean_value_y)
        denominator += (x - mean_value_x) ** 2
    if denominator == 0:
        return 0
    slope = numerator / denominator
    return slope
m = calculate_slope_ols(X, Y)
print(f"Slope = {m}")
```

```python
#intercept c
intercept_c = mean_value_y - (m * mean_value_x)
print(f"Intercept = {intercept_c}")
```
