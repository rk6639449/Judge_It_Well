**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
a = plt.figure()
plt.subplots(nrows = 1, ncols = 2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
a,(ax1,ax2) = plt.subplots(nrows = 1, ncols = 2)
y = x*2
ax1.plot(x,y, color = 'blue', lw = 2, ls = '--')
z = x**2
ax2.plot(x,z, color = 'red', lw = 1, ls = '--')
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
a, (ax1,ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (12,3))
y = x*2
ax1.plot(x,y, color = 'blue', lw = 2, ls = '--')
z = x**2
ax2.plot(x,z, color = 'red', lw = 1, ls = '--')
```

# #**Plots**

Plot the following plots using matplotlib, you many generate your own data or get it from the internet:

1. Histogram
2. Scatter Plot
3. Boxplot
4. Pie charts

(Bonus - Read up on the use-case of histograms in EDA and ML in general, they are VERY useful! Also, look up **Violin plots** if you have extra time)


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Histogram
'''Heights of 30 students (cm)
Write this as one column of numbers:

148, 150, 151, 152, 153, 155, 156, 157, 158, 160,
161, 162, 163, 163, 164, 165, 166, 168, 169, 170,
171, 172, 173, 175, 176, 178, 180, 181, 183, 185'''
x = np.array([148, 150, 151, 152, 153, 155, 156, 157, 158, 160,161, 162, 163, 163, 164, 165, 166, 168, 169, 170,171, 172, 173, 175, 176, 178, 180, 181, 183, 185])
plt.hist(x, bins = 5)
plt.title('Heights of Students')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()


# Scatter Plot
''' Biology: enzyme activity vs temperature
Reaction rate (arbitrary units) measured at different temperatures
Temperature (C): 10, 20, 30, 40, 50, 60
Rate: 5, 12, 20, 25, 18, 7 '''
x = np.array([10, 20, 30, 40 ,50, 60])
y = np.array([5, 12, 20, 25, 18, 7])
plt.xlabel('Temperature (C)')
plt.ylabel('Rate')
plt.title('Enzyme activity vs temperature')
colour1 = np.array(['r', 'b', 'g', 'hotpink', 'black', 'yellow'])
plt.scatter(x, y, c = colour1)
plt.show()


# Boxplot
'''Boxplot:
Test scores for two classes
Class A: 9, 12, 14, 15, 18, 19, 21, 23, 24, 27
Class B: 11, 13, 16, 20, 22, 25, 26, 28, 31, 37
'''
ClassA = [9, 12, 14, 15, 18, 19, 21, 23, 24, 27]
ClassB = [11, 13, 16, 20, 22, 25, 26, 28, 31, 37]
plt.boxplot([ClassA, ClassB], labels=['Class A', 'Class B'])
plt.title('Test scores for two classes')
plt.ylabel('Scores')
plt.show()


# Pie chart
''' Pie chart: monthly expenses
Students monthly budget
Rent: 30
Food: 25
Transport: 15
Entertainment: 10
Savings: 12
Others: 8 '''
y = np.array([30, 25, 15, 10, 12, 8])
mylabels = ['Rent', 'Food', 'Transport', 'Entertainment', 'Savings', 'Others']
mycolours = ['r', 'b', 'g', 'hotpink', 'm', 'yellow']
myexplode = [0, 0, 0, 0, 0.3, 0]
wedges, texts = plt.pie(y, labels = mylabels, colors = mycolours, shadow = True, explode = myexplode, startangle = 90)
plt.legend(wedges, mylabels, title = "Student's Monthly Budget", loc = "best")
plt.show()
```
